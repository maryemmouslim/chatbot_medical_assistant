```python
import pandas as pd 
from supabase import create_client
from dotenv import load_dotenv
import os
```

```python
load_dotenv(r"C:\Users\Marye\Desktop\lc\medical_assistant\.env")

url=os.getenv("supabase_url")
key=os.getenv("supabase_key")
client = create_client(url, key)

print("connected")
```

```python
df = pd.read_csv("data/medical_dataset_combined.csv")
print(df.shape)
df.drop_duplicates(inplace=True)
df.dropna(inplace=True)
print(df.shape)
```

```python
for i in range (0, len(df)):
    lot=df[i:i+1000]
    donnees=lot[["question","answer"]].to_dict(orient="records")
    client.table("medical_qa").insert(donnees).execute()
    print("donées envoyées ")
```
